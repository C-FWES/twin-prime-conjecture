```python
import math
import matplotlib.pyplot as plt
from numpy import *

def is_prime(n: int):
    if n == 2:
        return True
    if n % 2 == 0 or n <= 1:
        return False

    sqr = int(math.sqrt(n)) + 1

    for d in range(3, sqr, 2):
        if n % d == 0:
            return False
    return True

count = 0
twin_primes = []
for i in range(100000):
    if is_prime(i) and is_prime(i+2):
        twin_primes.append((i, i+2))
        count += 1
```

```python
print(count)
# print(twin_primes)
    
```

```python
x_axis_first = []
y_axis_first = []
for pair in twin_primes:
    pair_list = list(pair)
    x_axis_first.append(pair_list[0])
    y_axis_first.append(pair_list[1])
    
p = plt.figure()
p.set_figwidth(30)
p.set_figheight(30)
plt.plot(x_axis_first, y_axis_first, 'o')
plt.show()
```

```python
p = plt.figure()
p.set_figwidth(20)
p.set_figheight(20)
plt.plot(twin_primes[:])
plt.show()
```

```python
x_axis_third = []
y_axis_third = []
p = plt.figure()
p.set_figwidth(100)
p.set_figheight(100)
for n in range(100):
    for pair in twin_primes:
        avg = pair[0] + pair[1] / 2
        m = 6/avg
        x_axis_third.append((n/m) + 6 * m)
        y_axis_third.append(n)
plt.plot(x_axis_third, y_axis_third, 'r+')
```
